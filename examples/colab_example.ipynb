## Computing RSE Values in Colab

Use the `compute_rse_core` function with direct arguments:

```python
result_path = compute_rse_core(
    path="/content/RSE_Atlas/examples/files/rings.smi",  # Colab path
    isomer_engine='rdkit',    # 'rdkit' or 'omega' 
    gpu_idx=0                 # Use GPU 0, or None for CPU
)
```

**🔬 Process:**
1. 🔗 Break rings in molecules  
2. 🧪 Generate 3D conformers
3. ⚡ Calculate thermodynamic properties
4. 🎯 Compute RSE with neural networks
5. 💾 Save results as `.sdf` and `.csv`

````python
# Let's examine the example SMILES file
example_file = "/content/RSE_Atlas/examples/files/rings.smi"

print(f"📄 Example file: {example_file}")
print("📋 Contents:")
with open(example_file, 'r') as f:
    lines = f.readlines()
    for i, line in enumerate(lines, 1):
        print(f"  {i}: {line.strip()}")
    
print(f"\n🔬 Total molecules: {len(lines)}")
````

````python
# Example: Compute RSE for the ring molecules
# ⚠️ Note: This computation can take several minutes

smi_file_path = "/content/RSE_Atlas/examples/files/rings.smi"

print(f"🔬 Processing molecules from: {smi_file_path}")
print(f"⏳ This will take several minutes...")

# Uncomment the lines below to run the computation
result_path = compute_rse_core(
    path=smi_file_path,
    isomer_engine='rdkit',     # or 'omega' (requires OE_LICENSE)
    gpu_idx=0               # use None for CPU, 0 for GPU  
)

print(f"\n✅ RSE computation completed!")
print(f"📊 Results saved to: {result_path}")
````

## 📥 Download Results Locally

There are several ways to download your CSV results file for local analysis:

````python
# Method 1: Direct browser download using files.download()
from google.colab import files

if 'result_path' in locals():
    # Download the CSV file
    csv_file = result_path
    print(f"📊 Downloading CSV file: {csv_file}")
    files.download(csv_file)
    
    # Also download the SDF file if you want the 3D structures
    sdf_file = result_path.replace('.csv', '.sdf')
    print(f"🧬 Downloading SDF file: {sdf_file}")
    files.download(sdf_file)
    
    print("✅ Files downloaded! Check your browser's Downloads folder")
else:
    print("⚠️ Run the RSE computation first!")
````

````python
# Method 2: Display file path for manual download
if 'result_path' in locals():
    csv_file = result_path
    sdf_file = result_path.replace('.csv', '.sdf')
    
    print("📂 File locations in Colab:")
    print(f"   CSV: {csv_file}")
    print(f"   SDF: {sdf_file}")
    print("\n💡 To download manually:")
    print("   1. Click the folder icon (📁) in the left sidebar")
    print("   2. Navigate to the file location")
    print("   3. Right-click the file → Download")
else:
    print("⚠️ Run the RSE computation first!")
````
